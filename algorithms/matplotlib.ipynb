**Matplotlib** 

a Python library used for creating static, animated, and interactive visualizations. It is a versatile tool that allows users to generate a wide variety of plots, including line plots, scatter plots, bar charts, and histograms. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x_array = np.array([0,130])
y_array = np.array([0,240])
plt.plot(x_array,y_array)
plt.show()
```

```python
plt.plot(x_array,y_array,'o')
plt.show()
```

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

"""returns only points on the graph"""
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
"""it assumes default x points as only y points sare given"""
ypoints = np.array([3,5,7,11,12,5,7]) 
plt.plot(ypoints)
plt.show()
```

```python
ypoints = np.array([1,3,5,10,5,7,8])
plt.plot(ypoints,marker = '*')
plt.show()
```

```python
plt.plot(ypoints,marker = 'o',mec = 'r',markersize = 10)
plt.show()
```

*  **markersize** is used to change the size of marker points on the plot.
*  **mec** stands for marker edge color, here 'r' indicates red.
*  **mfc** stands for marker full color, here 'b' indicates blue color.

```python
plt.plot(ypoints,marker = '*',markersize = 20,mec = 'r',mfc = 'b')
plt.show()
```

```python
plt.plot(ypoints,ls = 'dashed')
plt.show()
```

* **ls** stands for line style
* **linewidth** stands for width of the plotted line

```python
plt.plot(ypoints,ls = 'dashed',mec = 'b',mfc = 'r',markersize = 20,color ='c',marker = '*',linewidth = 5)
plt.show()
```

```python
plt.plot(ypoints,color = 'y',ls = 'dashed',linewidth = 10)
plt.show()
```

```python
x1 = np.array([0,1,2,3])
x2 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])
plt.plot(x1,y1,x2,y2)
plt.show()
```

By default shows different colors to differentiate the different plotted lines.

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.title("sports watch data")
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
plt.grid(ls = '--',linewidth = 1,color = 'green')
plt.show()
```

**Grid** function returns grids in the plot, if you want specific axis grid, provide axis = x, this returns grid x or vertical lines

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.scatter(x,y,color = "r") # scatters the points on the plot
x = np.array([80,85,90,95,100,105,110,117,120,125])
y = np.array([77,250,260,270,260,290,300,310,320,330])
plt.scatter(x,y,color = "g")
plt.show()
```

**Viridis** is color map library in python

```python
x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)
plt.scatter(x,y,c=z,cmap = 'viridis')
plt.show()
```
